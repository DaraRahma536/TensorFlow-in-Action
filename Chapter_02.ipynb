# **Chapter 2: TensorFlow 2**

# **1. Langkah Pertama dengan TensorFlow 2**
---
* TensorFlow 2 menggunakan execution style imperatif (eager execution) di mana operasi dieksekusi langsung saat didefinisikan.
* AutoGraph ```@tf.function``` digunakan untuk mengonversi kode Python menjadi data-flow graph secara otomatis, meningkatkan performa untuk operasi berulang.

Multilayer Perceptron (MLP) diimplementasikan dengan:
- ```tf.Variable``` untuk parameter (bobot dan bias) yang bisa berubah.
- ```tf.Tensor``` untuk output yang tetap.
- Operasi seperti ```tf.matmul``` untuk perkalian matriks dan ```tf.nn.sigmoid```/```tf.nn.softmax``` untuk aktivasi.

# **2. Blok Bangunan TensorFlow**
---
### **```1. tf.Variable```**
* Mutable – nilai dapat berubah setelah inisialisasi.
* Digunakan untuk parameter model (bobot, bias).
* Memiliki atribut: shape, initial value, dtype.

Contoh:

````python
w = tf.Variable(tf.keras.initializers.RandomNormal()(shape=[4,3]))
````

### **```2. tf.Tensor```**
* Immutable – nilai tetap setelah dibuat.
* Output dari operasi TensorFlow (misal: tf.add, tf.matmul).
* Dapat berupa skalar, vektor, matriks, atau tensor n-dimensi.

Contoh:

````python
t = tf.constant([[1,2],[3,4]])
````

### **```3. tf.Operation```**
* Operasi yang memanipulasi ```tf.Variable``` dan ```tf.Tensor```.
* Jenis operasi:  
-&nbsp;Aritmatika: ```tf.add```, ```tf.multiply```, dll.  
-&nbsp;Reduksi: ```tf.reduce_sum```, ```tf.reduce_mean```, dll.  
-&nbsp;Logika: ```tf.equal```, ```tf.less```, dll.  

Contoh reduksi:

````python
tf.reduce_mean(a, axis=1)  # rata-rata per kolom
````

# **3. Komputasi Terkait Jaringan Syaraf Tiruan (Neural Network)**
---
### **1. Perkalian Matriks (tf.matmul)**
* Digunakan dalam fully connected network.
* Contoh: mengonversi gambar RGB ke grayscale:

````python
grayscale = tf.matmul(rgb_image, weights)
````

### **2. Operasi Konvolusi (tf.nn.convolution)**
* Inti dari Convolutional Neural Networks (CNN).
* Digunakan untuk deteksi tepi, blur, dll.
* Filter (kernel) digeser di atas gambar, melakukan perkalian elemen dan penjumlahan.

Contoh deteksi tepi:

````python
edge = tf.nn.convolution(image, filter)
````

### **3. Operasi Pooling (```tf.nn.max_pool```, ```tf.nn.avg_pool```)**
* Digunakan untuk down-sampling (mengurangi ukuran feature map).
* Max pooling: mengambil nilai maksimum dalam window.
* Average pooling: mengambil nilai rata-rata dalam window.

Contoh:

````python
pooled = tf.nn.max_pool(input, window_size, strides, padding)
````
