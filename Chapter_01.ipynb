# **CHAPTER 1: The amazing world of TensorFlow**

# **1. Apa Itu TensorFlow?**
---
TensorFlow adalah framework machine learning end-to-end yang dikembangkan oleh Google, dirancang untuk mempercepat komputasi pada hardware teroptimasi seperti GPU dan TPU. TensorFlow menyediakan ekosistem lengkap mulai dari pengembangan model, pelatihan, evaluasi, hingga deployment. Keras, yang awalnya berdiri sendiri, sekarang telah terintegrasi sebagai submodule dalam TensorFlow dan menyediakan API tingkat tinggi untuk mempermudah pembangunan model neural network.

# **2. GPU vs CPU vs TPU**
---
* CPU seperti mobil: cepat, sedikit core, cocok untuk tugas kompleks dan beragam.
* GPU seperti bus: banyak core, optimal untuk komputasi paralel seperti perkalian matriks (banyak digunakan dalam deep learning).
* TPU seperti bus ekonomis: dirancang khusus untuk ML, menggunakan presisi lebih rendah (bfloat16), lebih efisien untuk inferensi skala besar.

# **3. Kapan dan Kapan Tidak Menggunakan TensorFlow**
---
### **✅ Kapan Menggunakan TensorFlow:**
* Prototyping model deep learning (CNN, RNN, LSTM, dll.)
* Membangun pipeline data besar dengan tf.data
* Deployment model di cloud dengan TensorFlow Serving
* Monitoring pelatihan model dengan TensorBoard
* Menjalankan model pada hardware teroptimasi (GPU/TPU)

### **❌ Kapan Tidak Menggunakan TensorFlow:**
* Untuk model machine learning tradisional (linear regression, decision tree) – lebih cocok pakai scikit-learn
* Manipulasi data terstruktur skala kecil – lebih efisien pakai pandas/NumPy
* Pipeline NLP kompleks (lemmatization, stemming) – lebih baik pakai spaCy

---
